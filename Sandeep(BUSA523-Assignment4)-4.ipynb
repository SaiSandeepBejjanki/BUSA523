```python
import pandas as pd
import matplotlib.pyplot as plt

asgndata=pd.read_csv("data.csv")
asgndata
```

# Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('default')
%matplotlib notebook

histogram = asgndata.CPI.plot.hist(bins=15, color="purple", ec="red", grid=False, figsize=(8,6), legend=True)
histogram.set_title("Histogram - CPI Distribution", color="black", fontsize=18)
histogram.set_xlabel("CPI Values", color="red", fontsize=12)
plt.show()

histogram.figure.savefig("Histogram.png")
```

# Histogram with multiple variables (Same Frame)

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('bmh')
%matplotlib notebook

histgm=asgndata[['MarkDown1', 'MarkDown2', 'MarkDown3']].plot.hist(color=['green', 'orange', 'blue'], ec="white", grid=True, legend=True, alpha=0.4)
histgm.set_title("Histogram with multiple variables in same chart", color="black", fontsize=18)
histgm.set_xlabel("CPI Values", color="red", fontsize=12)
plt.show()

histgm.figure.savefig("HistogramWithMV1.png")
```

# BarChart

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('default')
%matplotlib notebook

barchart=asgndata.plot.bar(x='Day type', y='Fuel_Price', color="olive", legend=True, figsize=(24,6))
plt.show()

barchart.figure.savefig("Barchart.png")
```

# LineChart

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
%matplotlib notebook

linechart=asgndata.plot.line(x='Date', y='dept_1 sales', color="magenta", legend=True, figsize=(24,6))
plt.show()
linechart.figure.savefig("Linechart.png")

```

# AreaChart

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('dark_background')
%matplotlib notebook

areachart=asgndata.plot.area(x='Date', y='Temperature', color="salmon", legend=True)
plt.show()

areachart.figure.savefig("Areachart.png")
```

# ScatterPlot

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib notebook

plt.axvline(65, c=(.5, .5, .5), ls='--')
plt.axhline(3.2, c=(.5, .5, .5), ls='--')
scatterplot= plt.scatter(asgndata.Temperature, asgndata.Fuel_Price, c=asgndata.Fuel_Price, cmap='Spectral', ec='k', s=30)
plt.colorbar(label='Fuel_Price')

plt.show()

scatterplot.figure.savefig("Scatterplot.png")
```
